# 01 Data Exploration
Explore SIR simulation data.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path('../data/processed/sir_vector_field.csv')
if data_path.exists():
    df = pd.read_csv(data_path)
    print(df.shape)
    print(df.describe())
else:
    print('Run generate_data.py first')
```

```python
# Plot S, I, R distributions
if data_path.exists():
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    for ax, col in zip(axes, ['S', 'I', 'R']):
        df[col].hist(ax=ax, bins=30)
        ax.set_title(col)
    plt.tight_layout()
    plt.show()
```
